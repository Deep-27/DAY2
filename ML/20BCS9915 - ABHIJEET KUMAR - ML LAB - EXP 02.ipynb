```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("tips.csv")
plt.scatter(data['day'], data['tip'])
plt.title("Scatter Plot")
plt.xlabel('Day')
plt.ylabel('Tip')
 
plt.show()
```

```python
plt.scatter(data['day'], data['tip'], c=data['size'],
            s=data['total_bill'])
plt.title("Scatter Plot")
plt.xlabel('Day')
plt.ylabel('Tip')
 
plt.colorbar()
 
plt.show()
```

```python
plt.plot(data['tip'])
plt.plot(data['size'])
plt.title("Scatter Plot")
plt.xlabel('Day')
plt.ylabel('Tip')
 
plt.show()
```

```python
plt.bar(data['day'], data['tip'])
plt.title("Bar Chart")
plt.xlabel('Day')
plt.ylabel('Tip')
 
plt.show()
```

```python
plt.hist(data['total_bill'])
 
plt.title("Histogram")
 
plt.show()
```

```python
data = pd.read_csv("tips.csv")
sns.scatterplot(x='day', y='tip', data=data,
               hue='sex')
plt.show()
```

```python
data = pd.read_csv("tips.csv")
sns.lineplot(data=data.drop(['total_bill'], axis=1))

plt.show()
```

```python
data = pd.read_csv("tips.csv")
sns.barplot(x='day',y='tip', data=data,
            hue='sex')
 
plt.show()
```

```python
data = pd.read_csv("tips.csv")
sns.histplot(x='total_bill', data=data, kde=True, hue='sex')
 
plt.show()
```
